```python
from deep_neural_network import neural_network
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def show_predictions(x1, x2, y_hat, title):
    y_to_show = list(y_hat[0,:])
    plt.scatter(x1, x2, c=y_to_show, s=40, cmap=plt.cm.Spectral);
    plt.title(title)
    plt.show()
    
def model(layer_dims, X, y, learning_rate=0.05):
    np.random.seed(1)
    nn = neural_network(layer_dims, learning_rate)
    costs = nn.train(X, y, print_costs=False)
    y_hat = nn.predict(X)
    neural_network.print_accuracy(y_hat, y, 'Overall')

    return y_hat
```

```python
np.random.seed(1)
X = np.random.randn(2,1000) * 3
X_squared = X **2

ones = np.random.randint(1,2,size=(1,2))
c = np.dot(ones,X_squared)
c = np.sqrt(c)
c.reshape(1,X_squared.shape[1])

y = c >= 3
y[y == True] = 1
y[y == False] = 0


print(X[0,:].shape)
print(X[1,:].shape)
print(len(y))
print(type(y))
#print(y)

show_predictions(X[0,:], X[1,:], y, "Actual")
```

```python
layer_dims = (2,1)
y_hat = model(layer_dims, X, y)

show_predictions(X[0,:],X[1,:],y_hat, "one layer model's (logistic regression) guess")
show_predictions(X[0,:],X[1,:],y, "actual")
```

```python
layer_dims = (2,20,1)
y_hat = model(layer_dims, X, y)

show_predictions(X[0,:],X[1,:],y_hat, "2 layer model's guess")
show_predictions(X[0,:],X[1,:],y, "actual")
```

```python
layer_dims = (2,20,5,1)
y_hat = model(layer_dims, X, y)

show_predictions(X[0,:],X[1,:],y_hat, "3 layer model's guess")
show_predictions(X[0,:],X[1,:],y, "actual")
```
